```python
# Your code goes here

import pandas as pd

url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv"

df_uno = pd.read_csv(url)

url_2 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv"
df_dos = pd.read_csv(url_2)

url_3 = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv"
df_tres = pd.read_csv(url_3)
```

```python
df_uno.head()
```

```python
df_dos.head()
```

```python
df_tres.head()
```

```python
df_total = pd.concat([df_uno, df_dos, df_tres], axis=0, join="outer")
df_total.head()
```

```python
df_total.columns
```

```python
df_total.columns = df_total.columns.str.lower().str.replace(" ", "_")
df_total.columns
```

```python
df_total.rename(columns={"st": "state"}, inplace=True)
df_total.columns
```

```python
df_total.head()
```

```python
def change_gender(gender):
    if gender == "Male":
        return str("M")
    elif gender == "female" or gender == "Femal":
        return str("F")
    else:
        return gender
```

```python
df_total["gender"] = df_total["gender"].apply(change_gender) 
```

```python
df_total["gender"].value_counts()
```

```python
df_total["customer_lifetime_value"] = df_total["customer_lifetime_value"].str.replace('%', '')
df_total.head()
```

```python
df_total["vehicle_class"].value_counts()
```

```python
def class_car (vehicle_class):
    if vehicle_class == str("Sports Car") or vehicle_class == str("Luxury SUV") or vehicle_class == str("Luxury Car"):
        return "Luxury"
    else:
        return vehicle_class

df_total["vehicle_class"] = df_total["vehicle_class"].apply(class_car)
```

```python
df_total["vehicle_class"].value_counts()
```

```python
df_total["number_of_open_complaints"] = df_total["number_of_open_complaints"].str.split("/").str[1]
```

```python
df_total['number_of_open_complaints'] = pd.to_numeric(df_total['number_of_open_complaints'])
df_total["number_of_open_complaints"].value_counts()
```

```python
df_total.isnull().sum()
```

```python
df_total.describe()
```

```python
df_total.describe(include="object")
```

```python
mean_income = df_total["income"].mean()

df_total["income"] = df_total["income"].fillna(mean_income)

mean_mpauto = df_total["monthly_premium_auto"].mean()

df_total["monthly_premium_auto"] = df_total["monthly_premium_auto"].fillna(mean_income)
```

```python
df_total.isnull().sum()
```

```python
df_total.dropna(inplace=True)
```

```python
df_total.isnull().sum()
```

```python
df_total.duplicated().sum()
```

```python
df_total.duplicated().any()
```

```python
df_total.drop_duplicates(inplace=True)
```

```python
df_total.reset_index(drop=True, inplace=True)
```

```python
df_total.duplicated().sum()
```

The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*
